# Convert Segmented Video To .mp4 Format

```python
def convert_avi_to_mp4(input_file: str, output_file: str):
    try:
        start_time = perf_counter()
        # Load the AVI file
        video = mp.VideoFileClip(input_file)

        # Convert the video to MP4 format
        video.write_videofile(output_file, codec="libx264")
        end_time = perf_counter()
        execution_time = end_time - start_time
        print("Conversion completed successfully.")
        print(f"Converting {input_file} to {output_file} takes {execution_time:.4f} seconds.")

    except FileNotFoundError:
        raise FileNotFoundError(f"Input file '{input_file}' does not exist.")
    except OSError as e:
        raise OSError(f"Error occurred during conversion: {e}")
```

```python
input_file = "../Tugas3-SegmentasiDanDeteksi/runs/segment/predict/new_york_street.avi"
output_file = "IBDA4311_202000138_StefannusChristian_SegmentasiDeteksi.mp4"

try:
    convert_avi_to_mp4(input_file, output_file)
except FileNotFoundError as e:
    print(f"Error: {e}")
except OSError as e:
    print(f"Error: {e}")
```

# Plot Segmented Results

```python
results_path = "./runs/segment/predict/new_york_street_frames/"

ncols = 3

def plot_segmented_frames(results_path: str, ncols: int):
    # read in all the images, which are all the same size
    files = sorted(listdir(results_path), key=lambda x: int(x.split('.')[0]))
    images = [plt.imread(f"{results_path}{file}") for file in files]

    nrows = (len(files) // ncols) + 1

    # get heights for images, the number must match the number for nrows
    heights = [im.shape[0] for im in images[:nrows]]

    # get widths for images, the number must match the number for ncols
    widths = [im.shape[1] for im in images[:ncols]]

    # set the figure width in inches
    fig_width = 15

    # calculate the figure width
    fig_height = fig_width * sum(heights) / sum(widths)

    # create the figure
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(fig_width, fig_height),
                             gridspec_kw={'wspace': 0.05, 'hspace': 0.1, 'left': 0, 'right': 1,
                                          'bottom': 0, 'top': 1, 'height_ratios': heights})

    # flatten the axis into a 1-d array to make it easier to access each axes
    axes = axes.flatten()

    # iterate through the axes and associated images
    for ax, image, file in zip(axes, images, files):
        # add the image to the axes
        ax.imshow(image)

        # remove ticks / labels
        ax.axis('off')

        # display image name as title
        ax.set_title(file, fontsize=8, pad=2)

    # Remove empty subplots
    for i in range(len(images), len(axes)):
        fig.delaxes(axes[i])

    plt.subplots_adjust(wspace=0.05, hspace=0.1, left=0, right=1, bottom=0, top=1)

    plt.show()

# plot_segmented_frames(results_path, ncols)
```
